```python
# Etape 1 : Monter Google Drive
from google.colab import drive
drive.mount("/content/drive")

import os

# Chemin vers votre dossier de donnees sur Drive
# Modifiez si votre dossier a un nom different
DATA_DIR = "/content/drive/MyDrive/AccidentsFrance"

print("Fichiers CSV trouves :")
try:
    files = sorted([f for f in os.listdir(DATA_DIR) if f.endswith(".csv")])
    for f in files:
        size = os.path.getsize(os.path.join(DATA_DIR, f)) / 1024 / 1024
        print(f"  {f} ({size:.1f} Mo)")
    print(f"Total : {len(files)} fichiers")
except FileNotFoundError:
    print(f"Dossier non trouve : {DATA_DIR}")
    print("Verifiez le nom de votre dossier sur Drive")
```

```python
# Etape 2 : Cloner le depot GitHub
# Remplacez par URL de VOTRE repo
GITHUB_REPO = "https://github.com/VOTRE_USER/VOTRE_REPO.git"
REPO_NAME   = "VOTRE_REPO"

if not os.path.exists(f"/content/{REPO_NAME}"):
    os.system(f"git clone {GITHUB_REPO} /content/{REPO_NAME}")
else:
    os.system(f"cd /content/{REPO_NAME} && git pull")

os.chdir(f"/content/{REPO_NAME}")
print("Repo pret :", os.getcwd())
```

```python
# Etape 3 : Installer les dependances
os.system("pip install streamlit plotly pandas numpy scipy statsmodels pyngrok -q")
print("Dependances installees")
```

```python
# Etape 4 : Lancer Streamlit
import subprocess, time
from pyngrok import ngrok

os.makedirs(".streamlit", exist_ok=True)
with open(".streamlit/config.toml", "w") as f:
    f.write("[server]
headless = true
port = 8501
enableCORS = false
")

proc = subprocess.Popen(
    ["streamlit", "run", "app.py",
     "--server.port=8501",
     "--server.headless=true",
     "--server.enableCORS=false"],
    stdout=subprocess.DEVNULL,
    stderr=subprocess.DEVNULL
)
time.sleep(6)

tunnel = ngrok.connect(8501)
print("=" * 55)
print("DASHBOARD ACCESSIBLE ICI :")
print(tunnel.public_url)
print("=" * 55)
print()
print("Le dashboard detecte automatiquement :", DATA_DIR)
print("Si besoin, modifiez le chemin dans la barre laterale.")
```

```python
# Optionnel : Redemarrer apres mise a jour GitHub
os.system("git pull")
proc.terminate()
time.sleep(2)
proc = subprocess.Popen(
    ["streamlit", "run", "app.py",
     "--server.port=8501",
     "--server.headless=true",
     "--server.enableCORS=false"],
    stdout=subprocess.DEVNULL,
    stderr=subprocess.DEVNULL
)
time.sleep(5)
print(f"Dashboard redémarre : {tunnel.public_url}")
```
